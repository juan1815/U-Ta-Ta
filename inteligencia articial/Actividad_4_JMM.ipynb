```python
# Authors: The scikit-learn developers
# SPDX-License-Identifier: BSD-3-Clause

import numpy as np

rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])

############################################

from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

regr_1 = DecisionTreeRegressor(max_depth=4)

regr_2 = AdaBoostRegressor(
    DecisionTreeRegressor(max_depth=4), n_estimators=300, random_state=rng
)

regr_1.fit(X, y)
regr_2.fit(X, y)

y_1 = regr_1.predict(X)
y_2 = regr_2.predict(X)

##################################################

import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette("colorblind")

plt.figure()
plt.scatter(X, y, color=colors[0], label="training samples")
plt.plot(X, y_1, color=colors[1], label="n_estimators=1", linewidth=2)
plt.plot(X, y_2, color=colors[2], label="n_estimators=300", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Boosted Decision Tree Regression")
plt.legend()
plt.show()
```

# **PARTE 2**

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
#
# El ejemplo que utilizo para este ejercicio es datos de una tienda de ropa de cuanta ropa se compra alguien y de que valor
# 1. Generamos algunos datos de ejemplo (clientes)
# 'make_blobs' crea grupitos de puntos para que el K-Means tenga algo que agrupar
X, y = make_blobs(n_samples=300,  # Queremos 300 clientes (puntos)
                  centers=4,    # Que se agrupen en 4 tipos de clientes (grupos)
                  cluster_std=0.60, # Qué tan "apretados" o "dispersos" están los clientes en cada grupo
                  random_state=0) # Para que los datos sean siempre los mismos cada vez que se ejecuta el cofigo

# 2. Aplicar el algoritmo K-Means para que nos agrupe los datos de los grupitos ya creados
kmeans = KMeans(n_clusters=4, random_state=0, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 3. Visualizar los resultados con nombres reales
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5, marker='X')

plt.title('Segmentación de Clientes por Gasto y Frecuencia')
plt.xlabel('Gasto Mensual Promedio en Ropa ($)')
plt.ylabel('Frecuencia de Compra (Veces al Mes)')
plt.grid(True)
plt.show()
```

Conclusión Teórica sobre Clustering:

El algoritmo de K-Means, por ejemplo, lo que hace es buscar el centro de esos grupos y asignar cada dato al grupo más cercano, como si fuera el iman que los atrae.
